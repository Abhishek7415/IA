```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('new_data.csv')
data_use = data[['price','area','Price_sqft']]
data_use
```

```python
#Splitting dataset for training and testing
train_data = data_use.iloc[0:500]
test_data = data_use.iloc[5700:5750]

# making target vaue and variables to do linear regression
X = train_data[['area']]
Y = train_data['price']

#Preparing model for linear regression
model = LinearRegression()
model.fit(X,Y)

#Prediction of values using linear regression model
Y_pred = model.predict(test_data[['area']])

#checking accuracy
r_squared = r2_score(test_data['price'], Y_pred)

# Visualize the data and the regression line
plt.scatter(test_data[['area']], test_data['price'], label='Actual data')
plt.plot(test_data[['area']], Y_pred, color='red', label='Linear regression line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression using machine learning model')
plt.legend()
plt.show()

# Display the slope, intercept, and R-squared value
print('Slope:', model.coef_)
print('Intercept:', model.intercept_)
print('R-squared:', r_squared)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
X = np.array(train_data['area'].iloc[0:500])
y = np.array(train_data['price'].iloc[0:500])

# Manual calculation of linear regression
n = len(X)
sum_X = np.sum(X)
sum_y = np.sum(y)
sum_X_squared = np.sum(X**2)
sum_Xy = np.sum(X * y)

# Calculate slope (m) and intercept (b)
slope = (n * sum_Xy - sum_X * sum_y) / (n * sum_X_squared - sum_X**2)
intercept = (sum_y - slope * sum_X) / n

# Create a linear regression line
regression_line = slope * np.array(test_data['area']) + intercept


# Manual calculation of R-squared
mean_actual = np.mean(test_data['price'])
ss_total = np.sum((test_data['price'] - mean_actual)**2)
ss_residual = np.sum((test_data['price'] - regression_line)**2)

# Calculate R-squared
r_squared = 1 - (ss_residual / ss_total)


# Visualize the data and the regression line
plt.scatter(test_data[['area']], test_data['price'], label='Actual data')
plt.plot(test_data[['area']], regression_line, color='red', label='Linear regression line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression using maths manual method')
plt.legend()
plt.show()


# Display the slope and intercept
print('Slope (Manual):', slope)
print('Intercept (Manual):', intercept)
print('R-squared (Manual):', r_squared)
```
