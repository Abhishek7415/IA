```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('new_data.csv')
data_use = data[['price','area','Price_sqft']]
data_use
```

```python
#Splitting dataset for training and testing
train_data = data_use.iloc[0:500]
test_data = data_use.iloc[5700:5750]

# making target vaue and variables to do linear regression
X = train_data[['area']]
Y = train_data['price']

#Preparing model for linear regression
model = LinearRegression()
model.fit(X,Y)

#Prediction of values using linear regression model
Y_pred = model.predict(test_data[['area']])

#checking accuracy
errors = (abs(test_data['price'] - Y_pred))
mae = np.mean(errors)
# print('Mean Absolute Error:', mae)


# Visualize the data and the regression line
plt.scatter(test_data[['area']], test_data['price'], label='Actual data')
plt.plot(test_data[['area']], Y_pred, color='red', label='Linear regression line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression using machine learning model')
plt.legend()
plt.show()

# Display the slope, intercept, and R-squared value
print('Slope:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean Absolute Error:', mae)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array(train_data['area'].iloc[0:300])
y = np.array(train_data['price'].iloc[0:300])

#Normalize the values
X = (X - np.mean(X)) / np.std(X)



# Hyperparameters
learning_rate = 0.001
epochs = 50

# Initialize parameters
m = 0.0  # Slope
c = 0.0  # Intercept

# Mean squared errors
MSE = []

# Perform gradient descent
for epoch in range(epochs):
    # Calculate predictions
    predictions = c + m * X

    # Calculate errors
    error = predictions - y
    squared_error = error**2  # Fix variable name
    mse = np.mean(squared_error)
    MSE.append(mse)
    
    # Calculation of error gradients
    dm = -2 * np.mean(error * X)
    dc = -2 * np.mean(error)

    # Update parameters using gradient descent
    m = m - learning_rate * dm
    c = c - learning_rate * dc
    
    # Print progress
    print(f'Epoch {epoch + 1}/{epochs}, MSE: {mse}')
    
    print(MSE)
    print(dm)
    print(dc)
    print(m)
    print(c)

# Create a linear regression line using the final parameters
regression_line = c + m * test_data['area']

# Visualize the data and the regression line
plt.scatter(test_data['area'], test_data['price'], label='Actual data')
plt.plot(test_data['area'], regression_line, color='red', label='Linear regression line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression using Manual Gradient Descent')
plt.legend()
plt.show()

# Display the final parameters
print('Intercept (c):', c)
print('Slope (m):', m)
```
