```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('new_data.csv')
data_use = data[['price','area','Price_sqft']]
data_use
```

```python
#Splitting dataset for training and testing
train_data = data_use.iloc[0:500]
test_data = data_use.iloc[5700:5750]

# making target vaue and variables to do linear regression
X = train_data[['area']]
Y = train_data['price']

#Preparing model for linear regression
model = LinearRegression()
model.fit(X,Y)

#Prediction of values using linear regression model
Y_pred = model.predict(test_data[['area']])

#checking accuracy
errors = (abs(test_data['price'] - Y_pred))
mae = np.mean(errors)
# print('Mean Absolute Error:', mae)


# Visualize the data and the regression line
plt.scatter(test_data[['area']], test_data['price'], label='Actual data')
plt.plot(test_data[['area']], Y_pred, color='red', label='Linear regression line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression using machine learning model')
plt.legend()
plt.show()

# Display the slope, intercept, and R-squared value
print('Slope:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean Absolute Error:', mae)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X, y, learning_rate=0.00001, epochs=1000, convergence_threshold=1e-6):
    """
    Perform gradient descent for linear regression with an intercept.

    Parameters:
    - X: Area
    - y: Final Price
    - learning_rate: Value of Alpha
    - epochs: Number of iterations (int)
    - convergence_threshold: Very small value

    Returns:
    - theta: Optimized parameters (contains intercept and slope)
    - cost_history: History of MSE values during optimization
    """
    m = len(y)  # Number of training examples
    n = X.shape[1]  # Number of features
    theta = np.zeros((n, 1))  # Initialize parameters with zeros
    cost_history = []

    for epoch in range(epochs):
        # Hypothesis
        h = np.dot(X, theta)

        # Calculate the error
        error = h - y

        # Calculate the gradient
        gradient = (2/m) * np.dot(X.T, error)

        # Update parameters
        theta = theta - learning_rate * gradient

        # Calculate the cost
        MSE = (1/(m)) * np.sum(error**2)
        cost_history.append(MSE)

        # Check for convergence
        if epoch > 0 and abs(cost_history[epoch-1] - MSE) < convergence_threshold:
            print(f'Converged at epoch {epoch}, MSE: {MSE}')
            break

        # Print the cost at every iteration
        if epoch % 1 == 0:
            print(f'Epoch {epoch}, MSE: {MSE}')

    return theta, cost_history


def plot_regression_line(X, y, theta):
    plt.scatter(X[:, 1], y, marker='o', label='Training Data')
    plt.plot(X[:, 1], np.dot(X, theta), color='red', label='Linear Regression Line')
    plt.xlabel('Area')
    plt.ylabel('Price')
    plt.title('Linear Regression with Gradient Descent')
    plt.legend()
    plt.show()

def plot_cost_history(cost_history):
    plt.plot(range(len(cost_history)), cost_history, color='blue')
    plt.xlabel('Epochs')
    plt.ylabel('MSE')
    plt.title('MSE History during Training')
    plt.show()

# Example usage with your data:
# Assuming you have the 'area' and 'price' columns in the 'train_data' DataFrame
X = np.array(train_data['area'].iloc[0:500]).reshape(-1, 1)
y = np.array(train_data['price'].iloc[0:500]).reshape(-1, 1)

# Add a bias term to the input features
X_b = np.c_[np.ones((len(X), 1)), X]

# Set hyperparameters
learning_rate = 0.0000001  # Reduced learning rate
epochs = 15

# Perform gradient descent
theta, cost_history = gradient_descent(X_b, y, learning_rate, epochs)

# Print the optimized parameters
print('Optimized Parameters:', theta)

X_t = test_data[['area']]
X_test = np.c_[np.ones((len(X_t), 1)), X_t]
Y_test = test_data['price']

# Plot the regression line
plot_regression_line(X_test, Y_test, theta)

# Plot the cost history
plot_cost_history(cost_history)

print(f'Equation of line is: y ={float(theta[0])} + {float(theta[1])}*x')
```
